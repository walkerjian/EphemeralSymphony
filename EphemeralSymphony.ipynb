### Polymorphism Overview

Polymorphism, from the Greek "poly" meaning many and "morph" meaning form, refers to the ability of a function, object, or operator to operate in different ways depending on the input types. In programming, polymorphism is a powerful concept used to enable flexibility and code reusability. There are three main types of polymorphism:

1. **Ad-hoc Polymorphism** (Function Overloading and Operator Overloading)
2. **Parametric Polymorphism** (Generics)
3. **Subtype Polymorphism** (Inheritance and Interfaces)

Let's dive deeper into each type and provide examples in different languages.

### 1. **Ad-hoc Polymorphism**
Ad-hoc polymorphism occurs when a function or operator is overloaded to work with different types or numbers of arguments. This is common in languages that support function overloading or operator overloading. The function behaves differently based on the specific type or number of arguments it receives.

#### Best Example: **C++ (Function Overloading)**

In C++, we can define multiple versions of a function with the same name but different parameter types. The appropriate function is chosen at compile time based on the argument types.

```cpp
#include <iostream>

class Math {
public:
    // Function for adding two integers
    int add(int a, int b) {
        return a + b;
    }
    
    // Overloaded function for adding two floating-point numbers
    double add(double a, double b) {
        return a + b;
    }
    
    // Overloaded function for concatenating two strings
    std::string add(std::string a, std::string b) {
        return a + b;
    }
};

int main() {
    Math math;
    
    // Calls the int version
    std::cout << "Int Addition: " << math.add(2, 3) << std::endl;
    
    // Calls the double version
    std::cout << "Double Addition: " << math.add(2.5, 3.5) << std::endl;
    
    // Calls the string version
    std::cout << "String Concatenation: " << math.add("Hello", " World") << std::endl;

    return 0;
}
```

#### Explanation:
- The `add` function is overloaded for three different data types: `int`, `double`, and `std::string`.
- Depending on the argument types passed, the compiler selects the appropriate function at compile time. This is **ad-hoc polymorphism** since the functions have different behaviors depending on the argument types.

---

### 2. **Parametric Polymorphism**
Parametric polymorphism refers to writing functions or data types that can operate on any type. In statically typed languages, this is often implemented using **generics**, allowing the function or class to be defined without specifying an exact type.

#### Best Example: **Haskell (Pure Parametric Polymorphism)**

Haskell is a strongly-typed functional language that supports parametric polymorphism naturally. Let's define a simple polymorphic function.

```haskell
-- A polymorphic function that works on any type
identity :: a -> a
identity x = x

-- A generic function that applies a function to two elements
applyTwice :: (a -> a) -> a -> a
applyTwice f x = f (f x)

main = do
    print (identity 5)         -- Works with Int
    print (identity "Hello")   -- Works with String
    print (applyTwice (*2) 10) -- Doubles a number twice
```

#### Explanation:
- The `identity` function works on any type `a`. It simply returns the input as the output.
- The `applyTwice` function takes a function `f` and applies it twice to a value `x`. The type variable `a` makes this function polymorphic; it works with any type that satisfies the function signature.
- In Haskell, this is a natural form of parametric polymorphism, as the function doesnâ€™t care about the actual type, as long as the types match in the usage context.

---

#### Alternative Example: **Java (Generics)**

Java uses **generics** to achieve parametric polymorphism. A class or method can operate on objects of various types while providing compile-time type safety.

```java
import java.util.ArrayList;

public class Box<T> {
    private T item;
    
    public void set(T item) {
        this.item = item;
    }
    
    public T get() {
        return this.item;
    }
    
    public static void main(String[] args) {
        Box<Integer> intBox = new Box<>();
        intBox.set(10);
        System.out.println("Integer Value: " + intBox.get());
        
        Box<String> strBox = new Box<>();
        strBox.set("Hello");
        System.out.println("String Value: " + strBox.get());
    }
}
```

#### Explanation:
- The `Box<T>` class is a generic class, meaning it can work with any type `T`.
- In `main`, we create `Box` objects for `Integer` and `String`. The same class can handle different types without modification, ensuring type safety at compile time.

### 3. **Subtype Polymorphism**
Subtype polymorphism occurs when a subclass can be substituted for its superclass. It is closely related to the principle of **inheritance** in object-oriented programming. The most common form is when a base class reference can point to objects of its subclasses, allowing for dynamic dispatch.

#### Best Example: **Java (Inheritance and Interfaces)**

In Java, subtype polymorphism is achieved through inheritance and interfaces. A base class reference can point to any subclass object, and the correct method is determined at runtime.

```java
// Superclass
class Animal {
    public void speak() {
        System.out.println("Animal speaks");
    }
}

// Subclass Dog
class Dog extends Animal {
    public void speak() {
        System.out.println("Dog barks");
    }
}

// Subclass Cat
class Cat extends Animal {
    public void speak() {
        System.out.println("Cat meows");
    }
}

public class Main {
    public static void main(String[] args) {
        Animal myAnimal = new Animal();  // Animal object
        Animal myDog = new Dog();        // Dog object
        Animal myCat = new Cat();        // Cat object
        
        myAnimal.speak(); // Outputs "Animal speaks"
        myDog.speak();    // Outputs "Dog barks"
        myCat.speak();    // Outputs "Cat meows"
    }
}
```

#### Explanation:
- `Dog` and `Cat` are subclasses of `Animal`, and they override the `speak()` method.
- At runtime, Java determines which `speak()` method to call based on the actual object type (`Dog` or `Cat`), even though the reference is of type `Animal`.
- This is **subtype polymorphism**, as it relies on the ability to treat objects of subclasses as instances of their parent class.

#### Alternative Example: **Python (Duck Typing)**

Python supports subtype polymorphism via **duck typing**. Instead of checking the class of an object, Python checks if it implements the required methods.

```python
class Dog:
    def speak(self):
        return "Woof"

class Cat:
    def speak(self):
        return "Meow"

def animal_sound(animal):
    print(animal.speak())

dog = Dog()
cat = Cat()

animal_sound(dog)  # Outputs "Woof"
animal_sound(cat)  # Outputs "Meow"
```

#### Explanation:
- Python doesnâ€™t require explicit type declarations. As long as an object implements the required method (`speak()`), it can be passed to the `animal_sound` function.
- This is a form of **subtype polymorphism** via duck typing. The function can work with any object that has a `speak()` method, regardless of the objectâ€™s class.
